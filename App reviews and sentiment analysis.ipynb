```python
hk_project = reviews_all( 'com.hikingproject.android', sleep_milliseconds=0, lang ='en',country='US', sort=Sort.NEWEST)
```

```python
hk_project
```

```python
df = p.json_normalize(hk_project)


df.head()
```

```python
from transformers import pipeline


sentiment_analysis = pipeline("sentiment-analysis",model="siebert/sentiment-roberta-large-english")
```

```python
df['content'] = df['content'].astype('str')


df['result'] = df['content'].apply(lambda x: sentiment_analysis(x))


df.head()
```

```python
df['sentiment']= df['result'].apply(lambda x: (x[0]['label']))

df['score']= df['result'].apply(lambda x: (x[0]['score']))
```

```python
fig=px.histogram(df,x='sentiment',color='sentiment',text_auto=True)
fig.show()
```
